```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def detect_peaks(x):
    # We get array of diffs between each neighboring numbers
    dx = x[1:] - x[:-1]
    # We find potential peaks indices, looking for points where the diff is greater than zero from by left and lower from zero by right. 
    indices = np.where((np.hstack((dx, 0)) < 0) & (np.hstack((0, dx)) > 0))[0]
    # We delete first and last indices - they surely won't be peaks.
    if indices[0] == [0]:
        indices = indices[1:]
    if indices[-1] == x.size-1:
        indices = indices[:-1]
    return indices
```

```python
x = np.random.randn(100) + 10
```

```python
peaks = detect_peaks(x)
```

```python
plt.plot(x)
plt.plot(peaks, x[peaks], 'ro')
```

```python
plt.show()
```
